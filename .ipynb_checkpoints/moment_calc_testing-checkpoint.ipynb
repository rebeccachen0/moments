```python
import numpy as np
import pylab as mplot
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.interpolate import interp1d
```

```python
C_ell = np.loadtxt('../demo6/shear_cl/bin_1_1.txt')
ell = np.loadtxt('../demo6/shear_cl/ell.txt')
```

```python
mplot.figure()
mplot.plot(ell,(ell*(ell+1)*C_ell)/2*np.pi, color='c')
mplot.show()
```

```python
f = interp1d(ell, C_ell)
int_ell = np.arange(np.min(ell), np.max(ell), 1)
plt.plot(int_ell, (int_ell*(int_ell+1)*f(int_ell))/2*np.pi)
plt.show()
```

```python
def W_ell(l, theta):
    return np.exp(-(l**2)*(theta**2)/4)
def moment(theta):
    return np.sum((2*int_ell+1)*f(int_ell)*(W_ell(int_ell, theta)**2))/(4*np.pi)
```

```python
theta = np.linspace(0, 2*np.pi*(1./360), num=100)
fake_theta = np.linspace(0, 2*np.pi*(1./360), num=10)
print(fake_theta)
for i in range(0, 10):
    print (round(moment(fake_theta[i]), 6))
```

```python
for i in range(0, 100):
    #print(i, theta[i]*180./np.pi*60, moment(theta[i]))
    plt.semilogy(theta[i]*180./np.pi*60, moment(theta[i]), marker = 'x')
plt.show()
```
